```python
import folium
import re
import pandas as pd
import numpy as np
```

```python

def extract_lat_long_from_url(url,coordinate_choice):
    """
    Extract latitude and longitude from a Google Maps URL.

    :url: Google Maps URL as a string.
    :return: A tuple containing latitude and longitude as strings, or (None, None) if not found.
    """
    regex = '(!3d[0-9]{2,}\.[0-9]{3,})(!4d[0-9]{2,}\.[0-9]{3,})'
    match = re.search(regex, url)
    
    if match:
        latitude, longitude = match.groups()
        latitude=latitude[3:len(latitude)]
        longitude=longitude[3:len(longitude)]
        if coordinate_choice==1:
            return latitude
        else:
            return longitude
    else:
        return None
```

```python
bus_stops= pd.read_csv("./bus_stops.csv")
bus_stops.dropna(inplace=True)
bus_stops.rename(columns={"hfpxzc href":"Links"},inplace=True)
bus_stops.rename(columns={"W4Efsd 3":"Etiket"},inplace=True)
bus_stops['Links'] = bus_stops['Links'].astype(str)
bus_stops
```

```python
#bus_stops[["Coordinates","2"]]=bus_stops["Links"].apply(extract_lat_long_from_url[0])
bus_stops = pd.DataFrame(bus_stops)
bus_stops['latitude'] = bus_stops['Links'].apply(extract_lat_long_from_url, coordinate_choice=1)
bus_stops['longitude'] = bus_stops['Links'].apply(extract_lat_long_from_url, coordinate_choice=2)
bus_stops
```
